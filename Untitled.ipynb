```python
import pandas as pd
from matplotlib import pyplot as plt
import sqlite3
import math 
import numpy as np
plt.style.use('fivethirtyeight')

from IPython.core.display import display, HTML
```

```python
conn = sqlite3.connect('databases.db')
```

```python
c = conn.cursor()
```

```python
df = pd.read_csv ('dataset.csv')
```

```python
c.executescript('drop table if exists laporan')
c.executescript('drop table if exists destinasi')
c.executescript('drop table if exists time')
c.execute("CREATE TABLE IF NOT EXISTS 'destinasi' ('iddestinasi' INT, 'daerah_tujuan' VARCHAR(255) NULL, PRIMARY KEY ('iddestinasi'))")
c.execute("CREATE TABLE IF NOT EXISTS 'time' ('idtime' INT, 'bulan' VARCHAR(255) NULL, PRIMARY KEY ('idtime'))")
c.execute("CREATE TABLE IF NOT EXISTS 'laporan' ( 'idfact' INT, 'idtime' INT, 'iddestinasi' INT, 'berat_total' INT, PRIMARY KEY ('idfact'), FOREIGN KEY ('idtime') REFERENCES 'time' ('idtime') ON DELETE NO ACTION ON UPDATE NO ACTION, FOREIGN KEY ('iddestinasi') REFERENCES 'destinasi' ('iddestinasi') ON DELETE NO ACTION ON UPDATE NO ACTION)")
```

```python
destinasi = []
time = []
laporan = []
for i,j in df.iterrows():
    if j[1] not in time:
        time.append(j[1])
        
    t = -1
    for index, value in enumerate(time):
        if value == j[1]: 
            t = index+1
            break
        
    if j[2] not in destinasi:
        destinasi.append(j[2])
        
        

laporan = []
for i,j in df.iterrows():   
    t = -1
    for index, value in enumerate(time):
        if value == j[1]:
            t = index+1
            break
    
    k = -1
    for index, value in enumerate(destinasi):
        if value == j[2]:
            
            k = index+1
            
    
    laporan.append([t, k, j[3]])
```

```python
c.execute("DELETE FROM TIME")
for index, i in enumerate(time):
    c.execute("INSERT INTO TIME VALUES(" + str(index+1) + ", '" + str(i) + "')")
    
c.execute("DELETE FROM DESTINASI")
for index, i in enumerate(destinasi): 
    c.execute("INSERT INTO DESTINASI VALUES(" + str(index+1) + ", '" + str(i) + "')")

c.execute("DELETE FROM LAPORAN")
for index, i in enumerate(laporan):
    idtime = i[0]
    iddestinasi = i[1]
    c.execute("INSERT INTO LAPORAN VALUES(" + str(index+1) + "," + str(idtime) + "," + str(iddestinasi) + ",'" + str(i[2]) + "')")
    
conn.commit()
```

```python
c.execute('''select * from
(
select idtime, sum(berat_total) as jumlah_total_beras from laporan group by idtime
)''')
jumlah = []
for row in c.fetchall():
    jumlah.append(row)
np_bulan_time = np.array(jumlah)

plt.plot([int (x[0]) for x in np_bulan_time [:,:]], [ int(x[1]) for x in np_bulan_time[:,:]],'-o',linewidth=1.0)
plt.xticks([ int(x[0]) for x in np_bulan_time[:,:]], [ int(x[0]) for x in np_bulan_time[:,:]])
plt.yticks([ int(x[1]) for x in np_bulan_time[:,:]], [ int(x[1]) for x in np_bulan_time[:,:]])
plt.xlabel('Bulan')
plt.ylabel('Berat Beras Total / Ton')
plt.title("Distribusi Beras Disemua Daerah Tujuan Pada 2012  ")
plt.show()

df = pd.DataFrame([ x for x in np_bulan_time[:,:]])
df.columns = ['Bulan', 'Berat Beras Total / Ton']
df.index = np.arange(1,len(df)+1)
df
```

```python
c.execute('''select * from
(
select iddestinasi , max(berat_total) as max_total_beras from laporan group by iddestinasi
)''')
total = []
for row in c.fetchall():
    total.append(row)
np_daerah_tujuan_destinasi = np.array(total)

plt.plot([int (x[0]) for x in np_daerah_tujuan_destinasi [:,:]], [ int(x[1]) for x in np_daerah_tujuan_destinasi[:,:]],'-o',linewidth=1.0)
plt.xticks([ int(x[0]) for x in np_daerah_tujuan_destinasi[:,:]], [ int(x[0]) for x in np_daerah_tujuan_destinasi[:,:]])
plt.yticks([ int(x[1]) for x in np_daerah_tujuan_destinasi[:,:]], [ int(x[1]) for x in np_daerah_tujuan_destinasi[:,:]])
plt.xlabel('daerah_tujuan')
plt.ylabel('Berat Beras / Ton')
plt.title("Distribusi Beras Paling Banyak Disemua Daerah Tujuan Pada 2012  ")
plt.show()

df = pd.DataFrame([ x for x in np_daerah_tujuan_destinasi[:,:]])
df.columns = ['daerah_tujuan', 'Berat Beras / Ton']
df.index = np.arange(1,len(df)+1)
df
```
